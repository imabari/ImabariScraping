# スクレイピング

```python
url = "https://www.pref.ehime.jp/h25115/kanjyo/topics/influ2223/index2223.html"
```

```python
df0 = pd.read_html(url, index_col=[0, 1], header=[0], skiprows=1, match="定点医療機関における保健所別患者報告数")[0]
df0.index = df0.index.droplevel(0)
df0
```

```python
ss = df0.loc["定点数"].astype(int)
ss
```

```python
df0.drop("定点数", inplace=True)
df0.reset_index(inplace=True)
df0.rename(columns={"index": "週"}, inplace=True)
df0
```

# 定点当たり報告数

```python
df1 = df0.drop_duplicates(subset=["週"], keep="first").set_index("週").fillna(0)
df1
```

```python
df1.plot()
```

```python
df2 = df0.drop_duplicates(subset=["週"], keep="last").set_index("週").fillna("")
df2
```
