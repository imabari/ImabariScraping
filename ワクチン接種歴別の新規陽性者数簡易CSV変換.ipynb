# 設定

```python
url = "https://www.mhlw.go.jp/content/10900000/000975388.pdf" #@param {type:"string"}
page_no = 2 #@param {type:"integer"}
view_date = "2022-07-31" #@param {type:"date"}
```

# ダウンロード

```python
!wget $url -O data.pdf
```

```python
import pdfplumber
import pandas as pd
```

```python
# ページ番号変換
page_num = lambda x: x - 1
```

```python
# 表抽出

with pdfplumber.open("data.pdf") as pdf:

    page = pdf.pages[page_num(page_no)]

    table = page.extract_table()
```

```python
# 列名

df_col = pd.DataFrame(table[:2]).fillna(method="ffill", axis=1).replace("\s", "", regex=True)
df_col.iat[0, 0] = view_date
df_col
```

```python
col = df_col.values.tolist()
```

```python
df = pd.DataFrame(table[2:], columns=col)
```

```python
df.set_index(df.columns[0])
```

```python
df
```

```python
df.to_csv("result.csv", encoding="utf_8_sig")
```
