```python
from lxml import etree
import pathlib
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
ns = {"tei": "http://www.tei-c.org/ns/1.0", "xml": "http://www.w3.org/XML/1998/namespace"}
```

```python
parser = etree.XMLParser(recover=True)
```

```python
def strjoin(s, w = " "):

    if s:
        result = w.join([" ".join([j for j in i.split() if j]) for i in s])

        return result
    
    else:
        return ""
```

# タイトルを一覧表示

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    title = strjoin(root.xpath("//tei:teiHeader//tei:title/text()", namespaces=ns), "/")

    result.append({"file": str(p), "title": title})
```

```python
df = pd.DataFrame(result)
df
```

# 綴じ方の情報を見てみる

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    for i in root.xpath("//tei:bindingDesc", namespaces=ns):

        binding = i.xpath("./tei:p/text()", namespaces=ns)[0]
        title = strjoin(i.xpath("//tei:title/text()", namespaces=ns), "/")

        result.append({"binding": binding, "title": title})
```

```python
df = pd.DataFrame(result)
df
```

# 書籍の形についての情報

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    for i in root.xpath("//tei:teiHeader/tei:fileDesc", namespaces=ns):

        title = strjoin(i.xpath(".//tei:title[@type='uniform']/text()", namespaces=ns))
        binds = strjoin(i.xpath(".//tei:bindingDesc/tei:p/text()", namespaces=ns))
        extent = strjoin(i.xpath(".//tei:extent/descendant::text()", namespaces=ns))

        result.append({"title": title, "binds": binds, "extent": extent})
```

```python
df = pd.DataFrame(result)
df
```

# 書籍の形状をカウントしてみる

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    extent = strjoin(root.xpath("//tei:physDesc//tei:extent/text()", namespaces=ns))

    result.append({"extent": extent})
```

```python
df0 = pd.DataFrame(result)
df0
```

```python
df1 = df0["extent"].str.extract(";(?P<size>.+)$",expand=True)
df1["size"] = df1["size"].str.strip()
```

```python
df1["size"].value_counts().reset_index()
```

# 来歴情報に登場する人名を見てみる

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    for i in root.xpath("//tei:teiHeader/tei:fileDesc", namespaces=ns):

        title = strjoin(i.xpath(".//tei:title/text()", namespaces=ns), "/")
        name = strjoin(i.xpath(".//tei:provenance//tei:persName[@type='standard']/text()", namespaces=ns), "/")

        if name:
            result.append({"name": name, "title": title})
```

```python
df = pd.DataFrame(result)
df
```

# 人ごとに、来歴として関与したタイトルをまとめて数えてみる

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    for i in root.xpath("//tei:teiHeader/tei:fileDesc", namespaces=ns):

        title = strjoin(i.xpath(".//tei:title[@type='uniform']/text()", namespaces=ns), "/")
        
        for j in i.xpath(".//tei:provenance//tei:persName[@type='standard']/text()", namespaces=ns):

            name = " ".join(j.split())
            result.append({"name": name, "title": title})
```

```python
df = pd.DataFrame(result)
df
```

```python
df["name"].value_counts()
```

# 人毎に、書名をリストしてみる

```python
df.groupby(["name"])["title"].apply(list)
```

# 人毎に、作成・刊行年をなるべくまとめてみる

```python
result = []

for p in pathlib.Path("cam_jp_xml").glob("*.xml"):

    tree = etree.parse(p, parser)
    root = tree.getroot()

    for i in root.xpath("//tei:teiHeader/tei:fileDesc", namespaces=ns):

        title = strjoin(i.xpath(".//tei:title[@type='uniform']/text()", namespaces=ns), "/")
        origdate = i.xpath(".//tei:history//tei:origDate", namespaces=ns)[0].get("from", "").strip()

        for j in i.xpath(".//tei:provenance//tei:persName[@type='standard']/text()", namespaces=ns):

            name = " ".join(j.split())
            result.append({"name": name, "title": title, "origdate": origdate})
```

```python
df0 = pd.DataFrame(result)
df0
```

```python
df1 = df0.groupby(["name", "origdate"])["origdate"].count().reset_index(name="count")
df1
```

```python
print(df1.to_csv())
```

# 人毎に、刊行年を世紀でまとめてみる

```python
df0 = pd.DataFrame(result)
df0
```

```python
df0["century"] = (((pd.to_numeric(df0["origdate"], errors="coerce") - 1) / 100) + 1).apply(np.floor).astype("Int64")
```

```python
flag = df0["century"].isna()
```

```python
df0["century"] = df0["century"].astype(str)
```

```python
df0["century"] = df0["century"].where(flag, df0["century"] + "世紀")
df0["century"] = df0["century"].mask(flag, "")
```

```python
df0
```

```python
df1 = df0.groupby(["name", "century"])["century"].count().reset_index(name="count")
df1
```

```python
print(df1.to_csv())
```
